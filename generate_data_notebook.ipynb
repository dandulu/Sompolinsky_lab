```python
IT_matrices=[]
IT_avgs=[]
V4_matrices=[]
V4_avgs=[]
i=0
j=0
for image_id in unique_ids:
    temp_df=neuronal_df[neuronal_df['image_id'] == image_id]
    reps=temp_df['repetition'].unique()
    V4_rows=[]
    V4_avg=[]
    IT_rows=[]
    IT_avg=[]
    print(image_id)
    print("large index: ",j)
    j+=1
    for rep in reps:
        #print("small index: ",i)
        i+=1
        rep_df=temp_df[temp_df['repetition']==rep]
        IT_response  = rep_df[rep_df['region'] == 'IT']
        V4_response = rep_df[rep_df['region'] == 'V4']
        IT_measured = IT_response['dicarlo.MajajHong2015'].to_list()
        V4_measured = V4_response['dicarlo.MajajHong2015'].to_list()
        V4_rows.append(V4_measured)
        IT_rows.append(IT_measured)
    print("small index: ",i)
    V4_arr=np.vstack(np.array(V4_rows))
    V4_avg=np.mean(V4_arr,axis=0)
    IT_arr=np.vstack(np.array(IT_rows))
    IT_avg=np.mean(IT_arr,axis=0)
    IT_matrices.append(IT_arr)
    IT_avgs.append(IT_avg)
    V4_matrices.append(V4_arr)
    V4_avgs.append(V4_avg)

np.save('IT_matrices.npy', np.array(IT_matrices, dtype=object), allow_pickle=True)
np.save('IT_avgs.npy', np.array(IT_avgs, dtype=object), allow_pickle=True)
np.save('V4_matrices.npy', np.array(V4_matrices, dtype=object), allow_pickle=True)
np.save('V4_avgs.npy', np.array(V4_avgs, dtype=object), allow_pickle=True)
```
